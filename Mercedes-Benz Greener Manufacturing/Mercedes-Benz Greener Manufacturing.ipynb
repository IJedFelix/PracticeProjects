```python
#removing unusable columns from train and test

usable_columns = list(set(train.columns)-set(["ID","y"]))

y_train = train["y"].values
id_test = test["ID"].values
print("y_train:",y_train)
print("id_test:",id_test)
print("y_train_shape:",y_train.shape)
print("id_test_shape:",id_test.shape)
```

## Dropping the  Zero variance columns from train and test data

```python
for column in usable_columns:
    cardinality = len(np.unique(X_train[column]))
    if cardinality ==1:
        X_train.drop(column,axis=1)
        X_test.drop(column,axis=1)
X_train.head()
```

## Label encoding to categorical variables in test an dtrain data

```python
for f in ["X0","X1","X2","X3","X4","X5","X6","X8"]:
    lbl = preprocessing.LabelEncoder()
    lbl.fit(list(X_train[f].values))
    X_train[f] = lbl.transform(list(X_train[f].values))
```

```python
X_train[["X0","X1","X2","X3","X4","X5","X6","X8"]].dtypes
```

```python
for g in ["X0","X1","X2","X3","X4","X5","X6","X8"]:
    lbl1 = preprocessing.LabelEncoder()
    lbl1.fit(list(X_test[g].values))
    X_test[g] = lbl1.transform(list(X_test[g].values))
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
pca1 = PCA()
pca1_final_train = pca1.fit_transform(X_train)
pca1_final_test = pca1.fit_transform(X_test)
```

```python
print(pca1.explained_variance_ratio_)
```

```python
n_comp = 12
pca = PCA(n_components = n_comp, random_state = 21)
pca2_result_train = pca.fit_transform(X_train)
pca2_result_test = pca.fit_transform(X_test)
```

```python
print(pca.explained_variance_ratio_)
```

## Predicting test_df values using XGBoost.

```python
import xgboost as xgb
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
X_train,X_valid,y_train,y_valid = train_test_split(pca2_result_train,y_train,test_size=0.20,random_state=21)
```

```python
d_train = xgb.DMatrix(X_train, label=y_train)
d_valid = xgb.DMatrix(X_valid, label=y_valid)
d_test = xgb.DMatrix(pca2_result_test)
```

```python
params = {}
params["objective"] = 'reg:linear'
params["eta"] = 0.02
params["max_depth"] = 6
params["subsample"] = 0.7
params["colsample_size"] = 0.7

def xgb_r2_score(preds,dtrain):
    labels = dtrain.get_label()
    return "r2",r2_score(labels,preds)

watchlist = [(d_train,"train"),(d_valid,"valid")]

clf =  xgb.train(params,d_train,1000,watchlist,early_stopping_rounds=50,feval =xgb_r2_score,maximize =True, verbose_eval=10)
```

```python
p_test = clf.predict(d_test)
p_test
```

## Creating a result dataframe

```python
sub = pd.DataFrame()
sub["ID"] = id_test
sub["y"] = p_test
sub.to_csv("xgb.csv",index=False)
```
